Requirements :

. Given n meetings ,your system design is capable of finding all the conflicting elements

.Your design should be in such a way that it is capable of finding out the maximum meetings in one room


>Simple Scheduler

.n meetings ,one day 

.If time slot avialable return Meeting is Booked

.If maximum meetings are done ,meeting can't be booked

.If meeting  slot is not avialable ask them to book another slot

```python
def scheduler(interval,intervals,max_meetings):#constant time complexity as no loops are used so O(1)
  duration = interval[1]-interval[0]#calculating the duration of meeting
  i = interval[0]
  if max_meetings <1:#if all time slots are finished
    print("Sorry! Meeting can't be scheduled as maximum meetings for today was reached")
    return 1
  elif (duration == 1)&(intervals[i] ==1):#if following time slot is available
    intervals[i]-=1#as time slot is used
    max_meetings-=1#as meeting was booked
    print("Your Time slot is booked.")
    return 1
  elif (duration == 2)&(intervals[i]==1)&(intervals[i+1]==1):#if following time slot is available
    intervals[i]-=1#as time slot is booked
    intervals[i+1]-=1#as time slot is booked
    max_meetings-=2#as meeting was booked
    print("Your Time slot is booked.")
    return 1
  elif (duration ==3)&(intervals[i]==1) &(intervals[i+1]==1)&(intervals[i+2]==1):#if following time slot is available
    intervals[i]-=1#as time slot is booked
    intervals[i+1]-=1#as time slot is booked
    intervals[i+2]-=1#as time slot is booked
    max_meetings-=3#as meeting was booked
    print("Your Time slot is booked")
    return 1
  else:#if time slot conflicts
    print("Your required Time slot is not available")
    print("Try a new time slot")
    return 0
def check_input_format(start,end):#to check whether the given input is correct or not
  if (start>end) or (start<0) or (start>23) or (end<1) or (end>24) or ((end-start)>3):#to check start ,end are in the given range or not
    return 0
  else:
    return 1
if __name__ == '__main__':
  day = [1]*24 #creating a day of 24 hrs represent by 24 -1's
  max_meetings = 24 #Maximum no of meetings if each meeting is 1 hr in 24hrs we have 24 meeting
  t = int(input("Enter the no of meetings: "))
  print("Enter the time in intergers only  ")
  print("Maximum meeting duration is 3hrs only")
  while t:#O(n) n= numbe of cases 
    t-=1
    try:#to handle expection occured when unable to convert given input into integers
      start,end = map(int,input("Enter the start time and time seperated by space like 4 5: ").split())
    except:
      print("your Input was not in integer format")
      t+=1
      continue
    if check_input_format(start,end)==1:
      if scheduler([start,end],day,max_meetings) == 1:#O(1) constant time complexity
        pass
      else:#as time slot conflicts we are giving another chance
        t+=1
        continue
    else:
      print("Your Input Format is not in the range or start_time>end_time or duration of meeting is greater than 3")
      t+=1
```
