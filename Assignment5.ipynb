```python
#Binary Search Node intially single node so left,right values are none
class BinarySearchNode:
  def __init__(self,data):
    self.left  = None
    self.right = None
    self.value = data
#Insert function to insert elements in Binary Search Tree
def insertBST(root,data):
  if root is None:#if the given tree is empty, Small Problem
    return BinarySearchNode(data)
  else:
    if root.value == data:#as duplicates are not allowed in Binary Search Tree
      return root
    elif root.value<data:# node>root so value goes to right
      root.right = insertBST(root.right,data)
    else:#node<root ,so value goes to left
      root.left = insertBST(root.left,data)
    return root
#Inorder Traversal
def inorder(root):
  if root:
    inorder(root.left)
    print(root.value,end = " ")
    inorder(root.right)
#Preorder Traversal
def preorder(root):
  if root:
    print(root.value,end=" ")
    preorder(root.left)
    preorder(root.right)
#Postorder Traversal
def postorder(root):
  if root:
    postorder(root.left)
    postorder(root.right)
    print(root.value,end=" ")
#finding max value
def find_max(root):
  temp = root
  while temp.right is not None:
    temp = temp.right
  return temp
#finding min value
def find_min(root):
  temp = root
  while temp.left is not None:
    temp = temp.left
  return temp
#counting nodes
def count_nodes(root):
  if root is None:
    return 0
  return 1+count_nodes(root.left)+count_nodes(root.right)
#deleting nodes using inorder succesor
def deleteNode1(root,data):
  if root is None:
    return None
  else:
    if data<root.value:
      root.left = deleteNode1(root.left,data)
    elif data>root.value:
      root.right = deleteNode1(root.right,data)
    else:
      if root.left is None and root.right is None:
        return None
      if root.left is None:
        return root.right
      if root.right is None:
        return root.left
      min_value = find_min(root.right)
      root.value = min_value.value
      root.right=deleteNode1(root.right,min_value.value)
  return root
#deleting nodes using inorder predecessor
def deleteNode2(root,data):
  if root is None:
    return None
  else:
    if data<root.value:
      root.left  = deleteNode2(root.left,data)
    elif data>root.value:
      root.right = deleteNode2(root.right,data)
    else:
      if root.left is None and root.right is None:
        return None
      if root.left is None:
        return root.right
      if root.right is None:
        return root.left
      max_value  = find_max(root.left)
      root.value = max_value.value
      root.left = deleteNode2(root.left,max_value.value)
  return root
    
r = BinarySearchNode(50)
r = insertBST(r,30)                             #                       50
r = insertBST(r,20)                             #                      /  \
r = insertBST(r,40)                             #                     30   70
r = insertBST(r,70)                             #                    / \   / \
r = insertBST(r,60)                             #                  20  40 60 80
r = insertBST(r,80)                             
print("Inorder   Traversal: ");inorder(r)
print("\n","Preorder  Traversal: ");preorder(r)
print("\n","Postorder Traversal: ");postorder(r)
print("\n")
print("Maximum value in Tree: ",find_max(r).value)
print("Minimum value in Tree: ",find_min(r).value)
print("Number of Nodes: ",count_nodes(r))
print("Deleting node 70 using Inorder Succesor:");deleteNode1(r,70)
print("Inorder traversal after deleting node 70:");inorder(r)
print("\n")
print("Deleting node 30 using Inorder predecessor: ");deleteNode2(r,30)
print("Inorder traversal after deleting node 30");inorder(r)
```
