```python
#Time complexity: O(n)
def arrayRotation(arr,n,k):#arr = array, n = length of array , k = amount of time to rotate
  if k>=n:#as array after n rotations remain same
    k = k%n#remainder
  for _ in range(k):#maximum k = n-1 .So Timecomplexity O(n-1)~O(n)
    arr = [arr[-1]]+arr#add last element to the starting of array,as each rotation last element goest first,first to second...till fourth to fifth
    arr.pop()#removing the duplicate last element
  return arr
#Driver code:
if __name__=='__main__':
  t = int(input())#testcases
  while t:
    t-=1
    n,k = list(map(int,input().split()))#inputting n,k
    arr   = list(map(int,input().split()))#inputting array
    arr1 = arrayRotation(arr,n,k)
    for i in arr1:#printing array
      print(i,end=" ")
```

```python
#TimeComplexit:O(n^2)+O(nlogn)~O(n^2)
def minimumDifference(arr,n,x):#arr = array ,n = len(array),x= number our target
  minimum = 99999 # intialize minimum to lowest 
  arr = sorted(arr)#O(nlogn)
  for i in range(n-2):#O((n-2)*(n-3)*) ~ O(n^2)
    pos1 = i+1#pointer 1
    pos2 = n-1#pointer 2
    while pos1<pos2:#O(n-3),two pointer method
      sum1 = arr[i]+arr[pos1]+arr[pos2]#sum of 3 elements
      if sum1 == x:#if sum ==x that means difference is zero 
        return 0#return zero
      elif abs(sum1-x)<minimum:#if sum1 <x and sum1-x < minimum
        minimum = x-sum1#update minimum
        pos1+=1#update pointer 1 to get better minimum
      else:#if sum1>x 
        pos2-=1#reduce pointer 2 by 1
  return minimum
#DriverCode
if __name__ == '__main__':
  t =int(input())#testcases
  while t:
    t-=1
    n = int(input())#n = length of array
    array = list(map(int,input().split()))
    x = int(input())#target
    print("Minmum Difference is : ",minimumDifference(array,n,x))
```

```python
#TimeComplexit : O(n) length of array 
def maximumSubArray(arr,n):#arr = array ,n = length of array
  max_global = max_current = arr[0]
  for i in range(1,n):#Time compleixty O(n-1)~O(n)
    max_current = max(arr[i],max_current+arr[i])#as arr[i]> max so far then max so far = arr[i]
    if max_current>max_global:#if max so far > then highest max 
      max_global = max_current#change highest max value to max so far
  return max_global
#Driver Code:
if __name__ == '__main__':
  t= int(input())#testcases
  while t:
    t-=1
    n = int(input())#inputting
    arr = list(map(int,input().split()))
    print("Maximum Sub Array Sum is :",maximumSubArray(arr,n))

```
