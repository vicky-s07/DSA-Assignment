Problem 1:Given a sorted array and Find arr[i]=i

Time complexity : O(logn)

Same as binary search 
Here the given array is sorted is there exists arr[i]=i then arr[i]-i will be less than zero for the left side and > 0 for the right side.

We will search for element arr[i]-i = 0


```python
def mBinarySearch(arr,low,high):
  if(low==high):
    if(arr[low]-low==0):#O(1)
      return low
    else:
      return -1
  else:
    mid = low + (high-low)//2
    if ((arr[mid]-mid)==0):
      return mid
    elif (arr[mid]-mid<0):
      return mBinarySearch(arr,mid+1,high)#T(n/2)
    else:
      return mBinarySearch(arr,low,mid-1)#T(n/2)
```

Problem 2: two Sum
Time complexity = O(n)


```python
def twoSum(nums,target):
  dict1={}
  for i,v in enumerate(nums):#O(n)
    remaining = target -v
    if remaining in dict1:#o(1)
      return (dict1[remaining],i)
    dict1[v]=i
  return []
```

```python
twoSum([2,7,11,15],9)
```

```python
twoSum([3,2,4],6)
```

```python
twoSum([3,3],6)
```
