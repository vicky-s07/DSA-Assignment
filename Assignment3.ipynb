Question : Count the number of inversions in the given array .

inversion occur if i < j and arr[ i ] > arr[ j ] where i , j are indexes of array


Logic : We use modified merge sort where left and right recursive call return inverse count and array rather than only array

Time complexity : O(nlogn) 

```python
def mergeSort(arr):# arr = array
  inv_count = 0 # intialization of inv_count = 0 for each iteration
  if len(arr)==1:#small problem if single element in array return zero,Time complexity : O(1)
    return 0
  if len(arr)>1:
    mid = len(arr)//2  # finding mid value to divide array 
    L = arr[:mid]      #Left array (left  subtree)
    R = arr[mid:]      #Right array(right subtree)
    inv_count+=mergeSort(L)  # Left  recursive call T(n/2) here recursive call modifies array and return inverse count and we add it recursively
    inv_count+=mergeSort(R)  # Right recursive call T(n/2) here recursive call modifies array and return inverse count and we add it recursively
    # merge procedure Timecomplexity: O(n)
    i = j = k= 0 #intializing two pointers i,j=0 and for merged array index k =0

    while i <len(L) and j < len(R) : # run the loop until one half is empty
      if L[i]<R[j]:# if L[i]<R[j] we just need to copy as inversion dosen't occur here
        arr[k]= L[i]#copying to merged array
        i+=1 #increment left pointer
      else:#inversion occur 
        arr[k]=R[j]#copying to merged array
        inv_count+=(len(L)-i)# count the number of inversion as left and right are sorted if L[i]>R[j]
                            #then rest of elements upto mid are greater than R[j] so no of inversion are Len(L)-i
        j+=1#incrementing right pointer
      k+=1#increming merged array index
    #checking if any elements present in L(left array) and copying them to merged array if any
    while i <len(L):
      arr[k]=L[i]
      i+=1
      k+=1
    #checking if any elements present in R(right array)and copying them to merged array if any
    while j < len(R):
      arr[k]=R[j]
      j+=1
      k+=1
  return inv_count#return there respective inverse count in each iteration

#Recursive relation T(n)=O(1)+2T(n/2)+O(n){for merging}
#By masters theorem time complexity is O(nlogn) better than brute force O(n^2)
```

```python
array1 = [70,50,60,10,20,30,80,15]
mergeSort(array1)
```

```python
array2 = [1,20,6,4,5]
mergeSort(array2)
```

***Problem 1***

Write a program that accepts n,k and finds the output n^1/k? Program should run in time ployinomial of logn 

In the first line consists of the testcases t>=1. Each subsequent line consistns of pairs n,k separated by a space.

```python
#Modified Binary Search 
def root(x,k):
  if k ==1:#small problem
    return x
  start = 1
  end = x
  while (start<=end):
    mid = (start+end)//2#dividing the interval for each iteration it is divided by half
    #constant time complexity O(1)
    if mid**k == x:
      return mid
    if mid**k <x:
      start = mid+1
      ans = mid#finally we reach it's approx there start<=end condition fails and return ans
    else:
      end = mid-1
  return ans
'''
Time complexity :T (n) = T(n/2)+ O(1) by substitution method we can get logn as time complexity
'''
if __name__ =='__main__':
  t= int(input())
  while t:
    t-=1
    a=list(map(int,input().split()))
    print(root(a[0],a[1]))
```
